```python
from pathlib import Path
import os
from sklearn.model_selection import train_test_split
import shutil
from tqdm import tqdm

#原始資料路徑
data_path = '/home/u7412932/multiple_car/yolov7/pig_heatstress_2024_shuffle/train'
data_type = ['images', 'labels']
image_list = [x.resolve() for x in Path(data_path).glob('images/*.jpg')]
print(len(image_list))

train_type = ['train', 'test', 'valid']
#存放資料夾
save_path = '/home/u7412932/multiple_car/yolov7/pig_heatstress_2024_shuffle2'
for i in train_type:
  for j in data_type:
    os.makedirs(os.path.join(save_path, i, j))


train_list, test_list = train_test_split(image_list, test_size=0.1, random_state=1)
train_list, valid_list = train_test_split(train_list, test_size=0.1, random_state=1)
print(len(train_list), len(test_list), len(valid_list))

dir_dic = {
    'train':train_list,
    'test':test_list,
    'valid':valid_list
}



for name in tqdm(list(dir_dic.keys())):
  paths = dir_dic[name]
  for path in paths:
    # image
    file_name = path.name
    shutil.copy2(str(path), f'/home/u7412932/multiple_car/yolov7/pig_heatstress_2024_shuffle2/{name}/images')
    # label
    label_path = str(path).replace('images', 'labels').replace('.jpg', '.txt')
    shutil.copy2(label_path, f'/home/u7412932/multiple_car/yolov7/pig_heatstress_2024_shuffle2/{name}/labels')
    with open(f'/home/u7412932/multiple_car/yolov7/pig_heatstress_2024_shuffle2/{name}.txt', 'a') as file:
      file.write(str(path))
      file.write('\n')
```
